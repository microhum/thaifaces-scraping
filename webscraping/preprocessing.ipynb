```python
# I use this notebook to sort datasets which failed to download in automation webscraping phrase.
# These are link saved in log files.
# I sort it and then download it by hands.

import pandas as pd
```

```python
log_demo = "scrape_files/thaifaces/log_demo_other.txt"
log_other = "scrape_files/thaifaces/log_other.txt"
log = "scrape_files/thaifaces/log.txt"
```

```python
def read_log(path):
    # Read the text file line by line
    with open(path, 'r') as file:
        lines = file.readlines()

    # Initialize lists to store data
    data = []

    # Process each line
    for line in lines:
        # Split the line by commas
        parts = line.strip().split(', ')
        # Convert the first two parts to integers
        parts = [int(parts[0]), int(parts[1])] + parts[2:]
        # Append the parts to the data list
        data.append(parts)

    # Create a pandas dataframe
    df = pd.DataFrame(data, columns=['page', 'num_font', 'Link 1', 'Link 2'])
    df = df.drop_duplicates()
    df = df.sort_values(by=['page', 'num_font'])
    return df
```

```python
df_demo = read_log(log_demo)
print(len(df_demo))
df_demo.head(10)
```

```python
filtered_df = df_demo[~df_demo['Link 2'].str.startswith('https://www.jipatype.com') & ~df_demo['Link 2'].str.startswith('https://www.f0nt.com')]
print(len(filtered_df))
print(filtered_df.nunique())
filtered_df["Link 2"].unique()
```

```python
filtered_df_none = filtered_df[filtered_df["Link 2"] == "None"]
print(len(filtered_df_none))
print(filtered_df_none.nunique())
filtered_df_none
```
