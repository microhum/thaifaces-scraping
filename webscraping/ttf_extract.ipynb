```python
# This notebook used for extracting all ttf format from all scraped compressed files (zip / rar).

import os
import shutil
import zipfile

font_paths = []
for root, dirs, files in os.walk("scrape_files"):
    for file_ in files:
      if file_[-3:] in ["ttf", "zip", "otf", "rar"]:
        file_path = os.path.join(root, file_)
        font_paths.append(file_path)
```

```python
font_paths
```

```python
os.makedirs("scrape_files/all_fonts_pre", exist_ok=True)
for path in font_paths:
    try: 
        if path[-3:] in ["zip"]:
            zip_ref = zipfile.ZipFile(path,'r')
            zip_ref.extractall("scrape_files/all_fonts_pre/")
    except Exception as e:
        print(e)
```

```python
# Extract .ttf
os.makedirs("scrape_files/all_fonts", exist_ok=True)
os.makedirs("scrape_files/all_fonts/ttf", exist_ok=True)
os.makedirs("scrape_files/all_fonts/otf", exist_ok=True)
os.makedirs("scrape_files/all_fonts/rar", exist_ok=True)
os.makedirs("scrape_files/all_fonts/zip", exist_ok=True)


font_paths_all = []
for root, dirs, files in os.walk("scrape_files/all_fonts_pre"):
    for file_ in files:
      file_path = root+"/"+file_

      if file_[-3:] == "ttf":
          shutil.copy(file_path, f"scrape_files/all_fonts/ttf/{file_}")
      elif file_[-3:] == "otf":
          shutil.copy(file_path, f"scrape_files/all_fonts/otf/{file_}")
      elif file_[-3:] == "rar":
          shutil.copy(file_path, f"scrape_files/all_fonts/rar/{file_}")
      elif file_[-3:] == "zip":
          shutil.copy(file_path, f"scrape_files/all_fonts/zip/{file_}")

# Get .ttf left 
for path in font_paths:
   if path[-3:] == "ttf":
      shutil.copy(path, "scrape_files/all_fonts/ttf/")
   elif path[-3:] == "otf":
      shutil.copy(path, "scrape_files/all_fonts/otf/")
   elif path[-3:] == "rar":
      shutil.copy(path, "scrape_files/all_fonts/rar/") # Ignore .zip
```
